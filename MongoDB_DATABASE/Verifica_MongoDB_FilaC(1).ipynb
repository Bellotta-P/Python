#3 gli ordini effettuati nel marzo del 2013


```python
import datetime
result = db.sales.find({"$and": [{"saleDate": {"$gte": datetime.datetime(2013,3,1)}} , {"saleDate": {"$lte": datetime.datetime(2013,3,31,23,59)} }]})
df = pd.DataFrame(list(result))
df
```

#4 l’età media dei clienti 


```python
result = db.sales.aggregate([
                             {"$group": {"_id": "-->", "età_media": {"$avg":"$customer.age"} } }                            
                            ])
df=pd.DataFrame(list(result))
df
```

#5 il numero di clienti uomini e donne


```python
result = db.sales.aggregate([
                             {"$group": {"_id" : "$customer.gender", "total" : {"$sum" : 1}}},
                             ])
df=pd.DataFrame(list(result))
df
```

#6 il numero di acquisti per ogni categoria. Ordinare le categorie in ordine decrescente


```python
result = db.sales.aggregate([
                             {"$unwind": "$items"},
                             {"$unwind": "$items.tags"},
                             {"$group": {"_id" : "$items.tags", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }}
                             ])
df=pd.DataFrame(list(result))
df
```

#7 il numero di acquisti per ogni negozio (‘storeLocation’)


```python
result = db.sales.aggregate([
                             {"$group": {"_id" : "$storeLocation", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }}
                             ])
df=pd.DataFrame(list(result))
df
```

#8 i negozi in cui sono stati effettuati meno di 100 acquisti


```python
result = db.sales.aggregate([
                             {"$group": {"_id" : "$storeLocation", "total" : {"$sum" : 1}}},
                             {"$match": { "total": { "$lt": 100 } } },
                             {"$sort": { "total": -1 }}
                             ])
df=pd.DataFrame(list(result))
df
```

#9 il negozio in cui sono stati effettuati meno acquisti

```python
result = db.sales.aggregate([
                             {"$group": {"_id" : "$storeLocation", "total" : {"$sum" : 1}}},
                             {"$match": { "total": { "$lt": 100 } } },
                             {"$sort": { "total": 1 }}
                             ])
df=pd.DataFrame(list(result))
df
```
