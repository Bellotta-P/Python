#Trovare il prodotto venduto

```python
result = db2.vendite.aggregate([
     
    { "$group": {
        "_id": "maxAge",
        "maxAge": {
            "$max": "$customer.age"
        },
        "customerdata": {
            "$push": {
                "age": "$customer.age",
                "email": "$customer.email"
            }
        }}
       },
       
      {'$unwind': '$customerdata'

     
    },
    {'$match':{'$expr':{"$eq":["$maxAge", "$customerdata.age"]}}}
    ])
df = pd.DataFrame(list(result))
df
```
