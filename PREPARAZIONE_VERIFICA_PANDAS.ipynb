```python
import pandas as pd
```

#Copiare il link, nonostante non funzioni e soprattutto bisogna stare attenti al formato

```python
df = pd.read_excel("https://github.com/wtitze/3E/blob/main/Medals.xlsx?raw=true")
df
```

```python
df[df["Year of release"] != 2020 ]
```

```python
df.drop('column_name', axis=1, inplace=True)
```

#ELIMINARE LE COLONNE RANK, TOTAL E RANK BY TOTAL

```python
df.drop(columns = ["Rank" , "Total" , "Rank by Total"],inplace = True)
df
```

#CAMBIARE IL NOME DELLA COLONNA TEAM/NOC IN TEAM

```python
df.rename(columns = {"Team/NOC" : "Team"},inplace = True)
df
```

#CREARE UNA NUOVA COLONNA DOVE FARE IL TOTALE DI MEDAGLIE DI OGNI NAZIONE

```python
df["Total"] = df["Gold"] + df["Silver"] + df["Bronze"]
df
```

#LE NAZIONI CHE HANNO PRESO PIù DI 80 MEDAGLIE

```python
df[df["Total"] > 80][["Team"]]
df[df["Total"] != 2020 ] [["Team"]]
```

#NUMERO MEDIO  DI MEDAGLIE VINTE DALLE NAZIONI

```python
df["Total"].mean()
```

#LE NAZIONI CHE HANNO VINTO PIU MEDAGLIE DELLA MEDIA


```python
df[df["Total"] > df["Total"].mean()][["Team" , "Total"]]
```

#*INIZIO ESERICIZI GROUPBY*

```python
df1 = pd.read_excel("https://github.com/wtitze/3E/blob/main/Athletes.xlsx?raw=true")
df1
```

#RAGGRUPPARE QUANTI ATLETI CI SONO

```python
df1.groupby("NOC").count()
```

 #QUANTI ATLETI CI SONO IN ITALIA

```python
df1[df1["NOC"].str.contains("Italy")].count()
```

#LE NAZIONI CHE HANNO PIU DI 300 ATLETi

```python
df2 = df1.groupby("NOC").count()
df2[df2["Discipline"] > 300]
```

#LA NAZIONE CHE HA PORTATO Pù ATLETI

```python
df2[df2["Name"] == df2["Name"].max()]
```

#IL NUMERO DI NAZIONI DI DISCIPLINA

```python
df2 = df1.groupby("Discipline").count()
df2
```

#la disciplina con meno squadre

```python
df3= df1.groupby("Language").count()
df3[df3 == df3.min()]["Language"]
```
