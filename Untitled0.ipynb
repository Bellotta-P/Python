Scrivere un pogramma che date le lunghezze di tre lati, dica se è possibile costruire un triangolo con i lati di quella lunghezza.

```python
x, y, z= int(input("inserisci la x: ")),  int(input("inserisci la y: ")),  int(input("inserisci la z: "))
if x+y > z and x+z > y and z+y > x:
  print("si, è un triangolo")
else:
  print("no, non è un triangolo")
```
