```python
Squeeze_Melchiorre = Pista_Ciclabile_Melchiorre.geometry.squeeze()
```

# trovare tutti i quartieri attraversati da questa pista ciclabile

```python
Squeeze_Melchiorre_Union = Squeeze_Melchiorre.unary_union
Squeeze_Melchiorre_Union
```

```python
Quartieri_Melchiorre = Quartieri[Quartieri.intersects(Squeeze_Melchiorre_Union)]
Quartieri_Melchiorre
```

```python
ax = Quartieri_Melchiorre.to_crs(epsg=3857).plot(figsize=(20,20), linewidth=1, facecolor= "Y", edgecolor = "K", alpha=0.5)
Squeeze_Melchiorre.to_crs(epsg=3857).plot(ax=ax, linewidth=4, facecolor= "B", edgecolor = "Blue")
contextily.add_basemap(ax)
```
